```python
import autogen

config_list = [
    {
        "model": "huggingfaceh4/zephyr-7b-beta",
        #"api_type": "open_ai",
        "base_url": "https://openrouter.ai/api/v1",
        "api_key": "aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa"
    }
]
```

```python
llm_config = {"config_list": config_list, "seed": 42}

user_proxy =  autogen.UserProxyAgent(
    name = "user_proxy",
    system_message = """Reply TERMINATE if the task has been solved at full satisfaction. Otherwise reply CONTINUE or the reason why the task is not solved yet""",
    code_execution_config = {"work_dir": "code"},
    human_input_mode = "TERMINATE",
    max_consecutive_auto_reply = 10,
    is_termination_msg = lambda x: x.get("content", "").rstrip().endswith("TERMINATE")
)
```

```python
assistant = autogen.AssistantAgent(
    name = "Assistant",
    llm_config = llm_config,
    system_message = "You are an helpful assistant that responds with clear and concise answers",
    #is_termination_msg = lambda x: x.get("content", "").rstrip().endswith("TERMINATE")
)
```

```python
task1 = """
I want to write a textbook for beginners on a new coding library. How should the text book be structured so it is easy to understand?  
"""
```

```python
user_proxy.initiate_chat(
    assistant,
    message = task1,
)
```
